#Data Types and Structures Questions

1. What are data structures, and why are they important?
   - Data structure is a way of organizing, storing, and managing data in a computer so that the data can be accessed, modified, processed, and manipulated easily. When data is organized in a structured way, operations on the data become simple and efficient.

   - Data structure is important because it makes programs efficient, fast, and easy to manage, and it helps in accessing and manipulating data easily.

2. Explain the difference between mutable and immutable data types with examples
   - Mutability refers to the ability to change data after a value is assigned. Mutable data types allow operations such as insert, delete, update, or pop after creation. Examples include list, set, and dictionary.

   - Immutability refers to data that cannot be changed after it is created. Examples include tuple and string. String operations like replace, uppercase, or lowercase create a new value and do not change the original data.

3. What are the main differences between lists and tuples in Python?
   - The main difference between a list and a tuple in Python is mutability. A list is mutable, which means its data can be changed after creation, so it is used when modification of data is allowed. Lists support operations such as insert and delete.

   - A tuple is immutable, which means its data cannot be modified once created. Tuples are used when data should not be changed in any condition and they do not support operations like insert or delete.

4. Describe how dictionaries store data.
   - A dictionary stores data in the form of key–value pairs. Each key is unique and is used to access its corresponding value. Dictionaries store data in an organized way, allowing fast access, insertion, updating, and deletion of values using keys.

5. Why might you use a set instead of a list in Python?
   - A set is used when only unique elements are required. Both set and list are mutable, but a set does not allow duplicate values and stores a unique collection of elements, whereas a list allows duplicate values and does not store unique elements. Therefore, sets are useful because they do not allow duplicates.

6. What is a string in Python, and how is it different from a list?
    - A string in Python is a sequence of characters used to store text and is immutable, meaning its value cannot be changed after creation. A list is a collection of elements that is mutable, so its values can be modified, inserted, or deleted.

7. How do tuples ensure data integrity in Python?
   - Tuples ensure data integrity in Python because they are immutable.
Once a tuple is created, its values cannot be changed, added, or deleted. This prevents accidental modification of data and helps keep the data safe, consistent, and reliable.

8. What is a hash table, and how does it relate to dictionaries in Python?
   - A hash table is a data structure that stores data using key–value pairs by converting keys into hash values for fast access. In Python, dictionaries are implemented using hash tables.

   - Dictionary keys must be immutable so their hash value remains stable. Therefore, data types like integers, strings, tuples, and frozensets can be used as keys, while mutable types like lists cannot.

9. Can lists contain different data types in Python?
   - Yes, lists can contain different data types in Python because a list can store integers, floats, strings, booleans, and even other lists together.

10. Explain why strings are immutable in Python?
    - Strings are immutable in Python so that their values cannot be changed after creation. This improves performance, ensures data integrity, and allows strings to be safely used as dictionary keys.

11. What advantages do dictionaries offer over lists for certain tasks?
    - Dictionaries offer advantages over lists because they store data in key–value pairs, allowing direct and fast access using keys. This makes searching and retrieving data more efficient compared to lists, which access values using indexes. Dictionaries also support nested structures, making them suitable for storing related and structured data.

12. Describe a scenario where using a tuple would be preferable over a list?
    - A tuple is preferable over a list when the data should not be changed in any condition. For example, employee information such as phone numbers or addresses can be stored in a tuple to keep the data safe. Since tuples are immutable, the data cannot be modified, which helps maintain data integrity.

13. How do sets handle duplicate values in Python?
    - In Python, a set is an unordered data structure that does not allow duplicate values. If duplicate values are assigned to a set, they are automatically ignored, and only unique elements are stored.

14. How does the “in” keyword work differently for lists and dictionaries?
    - The 'in' keyword works differently for lists and dictionaries. In a list, the 'in' keyword checks whether a value is present in the list and returns 'True' if the value exists. In a dictionary, the 'in' keyword checks only the keys, not the values. If the given item is a key, it returns 'True'; if it is a value, it returns 'False'.

15. Can you modify the elements of a tuple? Explain why or why not.
    - No, we cannot modify the elements of a tuple because a tuple is immutable. Once a tuple is created, its elements cannot be updated or modified.

16. What is a nested dictionary, and give an example of its use case?
    - A nested dictionary is a dictionary in which the value of one or more keys is another dictionary. It is used to store structured or related data. For example, student names can be stored as keys, and their subject marks can be stored as another dictionary as values.
        
              students = {
                  "Ekta": {"math": 85, "science": 90},
                  "Rahul": {"math": 70, "science": 75}
                   }
     - This nested dictionary stores each student’s marks in different subjects inside another dictionary.

17. Describe the time complexity of accessing elements in a dictionary.
    - Accessing elements in a dictionary has an average time complexity of O(1), because dictionaries use a hash table, which allows fast lookup using keys.

18. In what situations are lists preferred over dictionaries?
    - Lists are preferred over dictionaries when key–value pairs are not required. They are used when we need to store a simple, ordered collection of values and access elements using indexes. Lists are suitable when only values are needed and no mapping between keys and values is required.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
    - Dictionaries are considered unordered because they store data using a hash table, where each key is mapped to a hash value rather than an index. Data is accessed using keys instead of indexes, so the order of elements is not important. This makes data retrieval very fast and efficient, even though the elements are not stored in a fixed order.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
     - The difference between a list and a dictionary in terms of data retrieval is that a list retrieves data using indexes, so elements are accessed based on their position and order. A dictionary retrieves data using keys, which are converted into hash values using a hash table. This allows very fast data access with an average time complexity of O(1). Therefore, dictionaries are efficient for key-based data retrieval, while lists are suitable for ordered data access.

```python
''' 29. Write a code to combine two dictionaries into one and print the result.'''

#create dictionaries
info = {'name': 'Payal', 'age': 20, 'course': 'Python'}

students = {
     "Ekta": {"math": 85, "science": 90},
     "Rahul": {"math": 70, "science": 75}}

#combine two dictionaries
info.update(students)

# print dictionary
print(f'combined two dictionaries : {info}')

#one more way to combine
combined = {**info, **students}
print(f'Other way combined dictionary : {combined}')

#one more way to combine
combined1 = info | students
print(f'Other one combined dictionaries {combined1}')
```
